```python
import matplotlib.pyplot as plt

class Income():

    def __init__(self, year, month, salary, bonus):
        self.__year = year
        self.__month = month
        self.__salary = salary
        self.__bonus = bonus
        self.income = {self.__year: {"January": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "February": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "March": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "April": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "May": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "June": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "July": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "August": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "September": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "October": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "November": {"salary": 0.0, "bonus": 0.0, "total": 0.0}, 
                                       "December": {"salary": 0.0, "bonus": 0.0, "total": 0.0},
                                       "Total": {"salary": 0.0, "bonus": 0.0, "total": 0.0}}}
        if salary != None:
            self.income[year][month]["salary"] = salary
        if bonus != None:
            self.income[year][month]["bonus"] = bonus
        if salary != None and bonus != None:
            self.income[year][month]["total"] = self.__salary + self.__bonus
        if salary != None:
            self.income[year]["Total"]["salary"] = self.__salary 
        if bonus != None:
            self.income[year]["Total"]["bonus"] = self.__bonus
        if salary != None and bonus != None:
            self.income[year]["Total"]["total"] = self.__salary + self.__bonus

    def __str__(self):
        display = ("************************************************************"
            + "\n {:s}".format("                      INCOME - " + str(self.__year)) 
            + "\n{a:<15s} {b:<15s} {c:<15s} {d:<15s}".format(a="MONTH", b="SALARY", c="BONUS", d="TOTAL") 
            + "\n" + "************************************************************\n")
        for val in self.income[self.__year].items():
            display += "{a:<15s} {b:<15.2f} {c:<15.2f} {d:<15.2f}".format(a=val[0], b=val[1]["salary"], c=val[1]["bonus"], d=val[1]["total"]) + "\n" 
        display += "************************************************************\n"
        return display

    def monthly_income(self):
        return "Total income in " + str(self.__month) + ": ${}".format(self.__salary + self.__bonus)
    
    def add_income(self, year, month, salary = 0, bonus = 0):
        self.income[year][month]["salary"] += salary
        self.income[year][month]["bonus"] += bonus
        self.income[year][month]["total"] += (salary + bonus)
        self.income[year]["Total"]["salary"] += salary
        self.income[year]["Total"]["bonus"] += bonus
        self.income[year]["Total"]["total"] += (salary + bonus)
        return "Income has been added"
    
    def update_salary(self, year, month, salary):
        temp = self.income[year][month]["salary"]
        self.income[year][month]["salary"] = salary
        self.income[year][month]["total"] += (salary - temp)
        self.income[year]["Total"]["salary"] += (salary - temp)
        self.income[year]["Total"]["total"] += (salary - temp)
        return "Salary has been updated"
    
    def update_bonus(self, year, month, bonus):
        temp = self.income[year][month]["bonus"]
        self.income[year][month]["bonus"] = bonus
        self.income[year][month]["total"] += (bonus - temp)
        self.income[year]["Total"]["bonus"] += (bonus - temp)
        self.income[year]["Total"]["total"] += (bonus - temp)
        return "Bonus has been updated"

class Expense():
    
    def __init__(self, year, month, rent, grocery, auto, medical, other):
        self.__year = year
        self.__month = month
        self.__rent = rent
        self.__grocery = grocery
        self.__auto = auto
        self.__medical = medical 
        self.__other = other
        self.expenses = {self.__year: {"January": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "February": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "March": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "April": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "May": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "June": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "July": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "August": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "September": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "October": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "November": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "December": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}, 
                                       "Total": {"rent": 0.0, "grocery": 0.0, "auto": 0.0, "medical": 0.0, "other": 0.0, "total": 0.0}}}
        if rent != None:
            self.expenses[year][month]["rent"] = rent
        if grocery != None:
            self.expenses[year][month]["grocery"] = grocery
        if auto != None:
            self.expenses[year][month]["auto"] = auto
        if medical != None:
            self.expenses[year][month]["medical"] = medical
        if other != None:
            self.expenses[year][month]["other"] = other 
        if rent != None and grocery != None and auto != None and medical != None and other != None:
            self.expenses[year][month]["total"] = self.__rent + self.__grocery + self.__auto + self.__medical + self.__other
            self.expenses[year]["Total"]["total"] = self.__rent + self.__grocery + self.__auto + self.__medical + self.__other

    def __str__(self):
        display = ("*************************************************************************************************************"
        + "\n {:s}".format("                                             EXPENSES - " + str(self.__year))
        + "\n{a:<15s} {b:<15s} {c:<15s} {d:<15s} {e:<15s} {f:<15s} {g:<15s}".format(a="MONTH", b="RENT", c="GROCERY", d="AUTO", e="MEDICAL", f="OTHER", g="TOTAL") 
        + "\n" + "*************************************************************************************************************\n")
        for val in self.expenses[self.__year].items():
            display += "{a:<15s} {b:<15.2f} {c:<15.2f} {d:<15.2f} {e:<15.2f} {f:<15.2f} {g:<15.2f}".format(a=val[0], b=val[1]["rent"], c=val[1]["grocery"], d=val[1]["auto"], e=val[1]["medical"], f=val[1]["other"], g=val[1]["total"]) + "\n" 
        display += "*************************************************************************************************************\n"
        return display

    def monthly_expenses(self):
        return "Total expenses in " + self.__month + ": ${}".format(self.__rent + self.__grocery + self.__auto + self.__medical + self.__other)
    
    def add_expenses(self, year, month, rent, grocery, auto, medical, other):
        self.expenses[year][month]["rent"] += rent
        self.expenses[year][month]["grocery"] += grocery
        self.expenses[year][month]["auto"] += auto
        self.expenses[year][month]["medical"] += medical
        self.expenses[year][month]["other"] += other
        self.expenses[year][month]["total"] += (rent + grocery + auto + medical + other)
        self.expenses[year]["Total"]["rent"] += rent
        self.expenses[year]["Total"]["grocery"] += grocery
        self.expenses[year]["Total"]["auto"] += auto
        self.expenses[year]["Total"]["medical"] += medical
        self.expenses[year]["Total"]["other"] += other
        self.expenses[year]["Total"]["total"] += (rent + grocery + auto + medical + other)
        return "Expenses has been added"
    
    def update_rent(self, year, month, rent):
        temp = self.expenses[year][month]["rent"] 
        self.expenses[year][month]["rent"] = rent
        self.expenses[year][month]["total"] += (rent - temp)
        self.expenses[year]["Total"]["rent"] += (rent - temp)
        self.expenses[year]["Total"]["total"] += (rent - temp)
        return "Rent has been updated"

    def update_grocery(self, year, month, grocery):
        temp = self.expenses[year][month]["grocery"] 
        self.expenses[year][month]["grocery"] = grocery
        self.expenses[year][month]["total"] += (grocery - temp)
        self.expenses[year]["Total"]["grocery"] += (grocery - temp)
        self.expenses[year]["Total"]["total"] += (grocery - temp)
        return "Grocery has been updated"
    
    def update_auto_expenses(self, year, month, auto):
        temp = self.expenses[year][month]["auto"] 
        self.expenses[year][month]["auto"] = auto
        self.expenses[year][month]["total"] += (auto - temp)
        self.expenses[year]["Total"]["auto"] += (auto - temp)
        self.expenses[year]["Total"]["total"] += (auto - temp)
        return "Auto expenses have been updated"
    
    def update_medical_expenses(self, year, month, medical):
        temp = self.expenses[year][month]["medical"] 
        self.expenses[year][month]["medical"] = medical
        self.expenses[year][month]["total"] += (medical - temp)
        self.expenses[year]["Total"]["medical"] += (medical - temp)
        self.expenses[year]["Total"]["total"] += (medical - temp)
        return "Medical expenses have been updated"

    def update_other_expenses(self, year, month, other):
        temp = self.expenses[year][month]["other"] 
        self.expenses[year][month]["other"] = other
        self.expenses[year][month]["total"] += (other - temp)
        self.expenses[year]["Total"]["other"] += (other - temp)
        self.expenses[year]["Total"]["total"] += (other - temp)
        return "Other expenses have been updated"

class Investment():

    def __init__(self, year, month, stocks, crypto, retirement_fund, realstate):
        self.__year = year
        self.__month = month
        self.__stocks = stocks 
        self.__crypto = crypto
        self.__retirement_fund = retirement_fund
        self.__realstate = realstate
        self.investments = {self.__year: {"January": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "February": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "March": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "April": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "May": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "June": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "July": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "August": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "September": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "October": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "November": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "December": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}, 
                                       "Total": {"stocks": 0.0, "crypto": 0.0, "retirement_fund": 0.0, "realstate": 0.0, "total": 0.0}}}
        if stocks != None:
            self.investments[year][month]["stocks"] = stocks
        if crypto != None:
            self.investments[year][month]["crypto"] = crypto
        if retirement_fund != None:
            self.investments[year][month]["retirement_fund"] = retirement_fund
        if realstate != None:
            self.investments[year][month]["realstate"] = realstate
        if stocks != None and crypto != None and retirement_fund != None and realstate != None:
            self.investments[year][month]["total"] = (stocks + crypto + retirement_fund + realstate)
            self.investments[year]["Total"]["stocks"] = stocks
            self.investments[year]["Total"]["crypto"] = crypto
            self.investments[year]["Total"]["retirement_fund"] = retirement_fund
            self.investments[year]["Total"]["realstate"] = realstate
            self.investments[year]["Total"]["total"] = (stocks + crypto + retirement_fund + realstate)

    
    def __str__(self):
        display = ("***********************************************************************************************"
        + "\n {:s}".format("                                   INVESTMENTS - " + str(self.__year))
        + "\n{a:<15s} {b:<15s} {c:<15s} {d:<20s} {e:<15s} {f:<15s}".format(a="MONTH", b="STOCKS", c="CRYPTO", d="RETIREMENT FUND", e="REALSTATE", f="TOTAL") 
        + "\n" + "***********************************************************************************************\n")
        for val in self.investments[self.__year].items():
            display += "{a:<15s} {b:<15.2f} {c:<15.2f} {d:<20.2f} {e:<15.2f} {f:<15.2f}".format(a=val[0], b=val[1]["stocks"], c=val[1]["crypto"], d=val[1]["retirement_fund"], e=val[1]["realstate"], f=val[1]["total"]) + "\n" 
        display += "***********************************************************************************************\n"
        return display

    def monthly_investment(self):
        return "Total investment in " + self.__month + ": ${}".format(self.__stocks + self.__crypto + self.__retirement_fund + self.__realstate)
    
    def add_investment(self, year, month, stocks, crypto, retirement_fund, realstate):
        self.investments[year][month]["stocks"] += stocks
        self.investments[year][month]["crypto"] += crypto
        self.investments[year][month]["retirement_fund"] += retirement_fund
        self.investments[year][month]["realstate"] += realstate
        self.investments[year][month]["total"] += (stocks + crypto + retirement_fund + realstate) 
        self.investments[year]["Total"]["stocks"] += stocks
        self.investments[year]["Total"]["crypto"] += crypto
        self.investments[year]["Total"]["retirement_fund"] += retirement_fund
        self.investments[year]["Total"]["realstate"] += realstate
        self.investments[year]["Total"]["total"] += (stocks + crypto + retirement_fund + realstate)
        return "Investments have been added"
    
    def update_stocks(self, year, month, stocks):
        temp = self.investments[year][month]["stocks"] 
        self.investments[year][month]["stocks"] = stocks
        self.investments[year][month]["total"] += (stocks - temp)
        self.investments[year]["Total"]["stocks"] += (stocks - temp)
        self.investments[year]["Total"]["total"] += (stocks - temp)
        return "Stocks have been updated"
    
    def update_crypto(self, year, month, crypto):
        temp = self.investments[year][month]["crypto"] 
        self.investments[year][month]["crypto"] = crypto
        self.investments[year][month]["total"] += (crypto - temp)
        self.investments[year]["Total"]["crypto"] += (crypto - temp)
        self.investments[year]["Total"]["total"] += (crypto - temp)
        return "Crypto investment has been updated"
    
    def update_retirement_fund(self, year, month, retirement_fund):
        temp = self.investments[year][month]["retirement_fund"] 
        self.investments[year][month]["retirement_fund"] = retirement_fund
        self.investments[year][month]["total"] += (retirement_fund - temp)
        self.investments[year]["Total"]["retirement_fund"] += (retirement_fund - temp)
        self.investments[year]["Total"]["total"] += (retirement_fund - temp)
        return "Retirement_fund has been updated"
    
    def update_realstate(self, year, month, realstate):
        temp = self.investments[year][month]["realstate"] 
        self.investments[year][month]["realstate"] = realstate
        self.investments[year][month]["total"] += (realstate - temp)
        self.investments[year]["Total"]["realstate"] += (realstate - temp)
        self.investments[year]["Total"]["total"] += (realstate - temp)
        return "Realstate investment has been updated"

class Budget(Income, Expense, Investment):

    def __init__(self, year):
        self.__year = year
        Income.__init__(self, year = self.__year, month = None, salary = None, bonus = None)
        Expense.__init__(self, year = self.__year, month = None, rent = None, grocery = None, auto = None, medical = None, other = None)
        Investment.__init__(self, year = self.__year, month = None, stocks = None, crypto = None, retirement_fund = None, realstate = None)
    
    def __str__(self):
        display = ("************************************************************"
            + "\n {:s}".format("                     INCOME - " + str(self.__year)) 
            + "\n{a:<15s} {b:<15s} {c:<15s} {d:<15s}".format(a="MONTH", b="SALARY", c="BONUS", d="TOTAL") 
            + "\n" + "************************************************************\n")
        for val in self.income[self.__year].items():
            display += "{a:<15s} {b:<15.2f} {c:<15.2f} {d:<15.2f}".format(a=val[0], b=val[1]["salary"], c=val[1]["bonus"], d=val[1]["total"]) + "\n" 
        display += "************************************************************\n"
        display += ("*************************************************************************************************************"
        + "\n {:s}".format("                                            EXPENSES - " + str(self.__year))
        + "\n{a:<15s} {b:<15s} {c:<15s} {d:<15s} {e:<15s} {f:<15s} {g:<15s}".format(a="MONTH", b="RENT", c="GROCERY", d="AUTO", e="MEDICAL", f="OTHER", g="TOTAL") 
        + "\n" + "*************************************************************************************************************\n")
        for val in self.expenses[self.__year].items():
            display += "{a:<15s} {b:<15.2f} {c:<15.2f} {d:<15.2f} {e:<15.2f} {f:<15.2f} {g:<15.2f}".format(a=val[0], b=val[1]["rent"], c=val[1]["grocery"], d=val[1]["auto"], e=val[1]["medical"], f=val[1]["other"], g=val[1]["total"]) + "\n" 
        display += "*************************************************************************************************************\n"
        display += ("***********************************************************************************************"
        + "\n {:s}".format("                                   INVESTMENTS - " + str(self.__year))
        + "\n{a:<15s} {b:<15s} {c:<15s} {d:<20s} {e:<15s} {f:<15s}".format(a="MONTH", b="STOCKS", c="CRYPTO", d="RETIREMENT FUND", e="REALSTATE", f="TOTAL") 
        + "\n" + "***********************************************************************************************\n")
        for val in self.investments[self.__year].items():
            display += "{a:<15s} {b:<15.2f} {c:<15.2f} {d:<20.2f} {e:<15.2f} {f:<15.2f}".format(a=val[0], b=val[1]["stocks"], c=val[1]["crypto"], d=val[1]["retirement_fund"], e=val[1]["realstate"], f=val[1]["total"]) + "\n" 
        display += "***********************************************************************************************\n"
        return display

    def cash_flow(self):
        cf_list = []
        in_list = []
        ex_list = []
        for val_income in self.income[self.__year].items():
            in_list.append(val_income[1]["total"])  
        for val_expense in self.expenses[self.__year].items():
            ex_list.append(val_expense[1]["total"])
        i = 0
        while i < len(in_list):
            cf_list.append(in_list[i]-ex_list[i])
            i += 1
        display = ("********************************************************************************************************************************************"
                    + "\n{:50s}".format("                                                          CASH FLOW - " + str(self.__year))
                    + "\n********************************************************************************************************************************************" 
                    + "\n{a:<10s} {b:<10s} {c:<10s} {d:<10s} {e:<10s} {f:<10s} {g:<10s} {h:<10s} {i:<10s} {j:<10s} {k:<10s} {l:<10s} {m:<10s}"
                        .format(a="January", b="February", c="March", d="April", e="May", f="June", g="July", h="August", i="September", j="October", k="November", l="December", m="Total") + "\n") 
        for val in cf_list:
            display += "{a:<11.0f}".format(a=val)
        display += "\n********************************************************************************************************************************************"
        print(display)
        fig = plt.figure()
        ax = fig.add_axes([0,0,1,1])
        x_ax = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
        y_ax = cf_list
        y_ax.pop()
        ax.bar(x_ax, y_ax)
        plt.show()
    
    def net_worth(self):
        nw_list = []
        in_list = []
        ex_list = []
        inv_list = []
        for val_income in self.income[self.__year].items():
            in_list.append(val_income[1]["total"])  
        for val_expense in self.expenses[self.__year].items():
            ex_list.append(val_expense[1]["total"])
        for val_investment in self.investments[self.__year].items():
            inv_list.append(val_investment[1]["total"])
        i = 0
        while i < len(in_list):
            if i == 0:
                nw_list.append(in_list[i] - ex_list[i] + inv_list[i])
            else:
                nw_list.append(in_list[i] - ex_list[i] + inv_list[i] + nw_list[i-1])
            i += 1
        nw_list.pop()
        display = ("***********************************************************************************************************************************"
                    + "\n{:50s}".format("                                                         NET WORTH - " + str(self.__year))
                    + "\n***********************************************************************************************************************************" 
                    + "\n{a:<10s} {b:<10s} {c:<10s} {d:<10s} {e:<10s} {f:<10s} {g:<10s} {h:<10s} {i:<10s} {j:<10s} {k:<10s} {l:<10s}"
                        .format(a="January", b="February", c="March", d="April", e="May", f="June", g="July", h="August", i="September", j="October", k="November", l="December") + "\n") 
        for val in nw_list:
            display += "{a:<11.0f}".format(a=val)
        display += "\n***********************************************************************************************************************************"
        print(display)
        fig = plt.figure()
        ax = fig.add_axes([0,0,1,1])
        x_ax = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
        y_ax = nw_list
        ax.bar(x_ax, y_ax)
        plt.show()
    
class budget_manager():

    def __init__(self):
        self.account_name = []
        self.accounts = int(input("How many accounts you will be managing?"))
        self.year = int(input("For which year you will be managing the budget"))
        i = 0
        while i < self.accounts:
            name = input("Enter the accunt holder name")
            self.account_name.append(name) 
            i += 1
        self.account_list =  [Budget(self.year) for i in self.account_name]

    def get_accounts(self):
        i = 0
        while i < len(self.account_list):
            print("                   ***Budget for "+ self.account_name[i] + "***")
            print(self.account_list[i])
            i += 1
    
    def fill_income(self):
        salary = float(input("What is the monthly salary"))
        bonus = float(input("What is the monthly bonus"))
        months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
        i = 0
        j = 0
        while i < self.accounts:
            while j < 12:
                self.account_list[i].add_income(self.year, months[j], salary, bonus)
                j += 1
            j = 0
            i += 1
    
    def fill_expenses(self):
        rent = float(input("What is the monthly rent"))
        grocery = float(input("What are monthly grocery expenses"))
        auto = float(input("What are monthly auto expenses"))
        medical = float(input("What are monthly medical expenses"))
        other = float(input("What are other monthly expenses"))
        months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
        i = 0
        j = 0
        while i < self.accounts:
            while j < 12:
                self.account_list[i].add_expenses(self.year, months[j], rent, grocery, auto, medical, other)
                j += 1
            j =0
            i += 1

    def fill_investments(self):
        stocks = float(input("What is monthly stock investment"))
        crypto = float(input("What is monthly crypto investment"))
        retirement_fund = float(input("What is monthly 401K contribution"))
        realstate = float(input("What is monthly investment in realstate"))
        months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
        i = 0
        j = 0
        while i < self.accounts:
            while j < 12:
                self.account_list[i].add_investment(self.year, months[j], stocks, crypto, retirement_fund, realstate)
                j += 1
            j =0
            i += 1
    
    def accounts_snapshot(self):
        i = 0
        j = 0
        while i < self.accounts:
            print("                                                      ***CASH FLOW for "+ self.account_name[i] + "***")
            self.account_list[i].cash_flow()
            print("                                                      ***NET WORTH of "+ self.account_name[i] + "***")
            self.account_list[i].net_worth()
            i += 1
    
    def change_an_entry(self):
        account = input("Which account you want to update")
        month = input("Which month you want to update")
        # update = input("Which category you want to update, Income, Expenses or Investments")
        value = input("Which parameter you want tp update")
        new_value = float(input("What is the new value"))
        name_index = self.account_name.index(account)
        name = self.account_list[name_index]
        if value[:3].lower() == "sal":
            name.update_salary(self.year, month, new_value)
        elif value[:3].lower() == "bon":
            name.update_bonus(self.year, month, new_value)
        elif value[:3].lower() == "ren":
            name.update_rent(self.year, month, new_value)
        elif value[:3].lower() == "gro":
            name.update_grocery(self.year, month, new_value)
        elif value[:3].lower() == "aut":
            name.update_auto_expenses(self.year, month, new_value)
        elif value[:3].lower() == "med":
            name.update_medical_expenses(self.year, month, new_value)
        elif value[:3].lower() == "oth":
            name.update_other_expenses(self.year, month, new_value)
        elif value[:3].lower() == "sto":
            name.update_stocks(self.year, month, new_value)
        elif value[:3].lower() == "cry":
            name.update_crypto(self.year, month, new_value)
        elif value[:3].lower() == "ret":
            name.update_retirement_fund(self.year, month, new_value)
        else:
            value[:3].lower() == "rea"
            name.update_realstate(self.year, month, new_value)
    
# a = Income(2021, "January", 1200, 1200)
# print(a.monthly_income())
# a.add_income(2021, "February", 2000, 1000)
# a.update_bonus(2021, "January", 1500)
# b = Expense(2021, "January", 3500, 2000, 1000, 500, 1000)
# b.add_expenses(2021, "December", 3000, 2000, 1100, 200, 600)
# print(b.monthly_expenses())
# c = Investment(2021, "February", 1000, 1000, 200, 500)
# c.add_investment(2021, "January", 2000, 10000, 1400, 50000)
# b.update_auto_expenses(2021, "March", 200)
# print(c.monthly_investment())
# print(a)
# print(b)
# print(c)
# c.update_crypto(2021, "May", 1600)
# print(c)
# d = Budget(2022)
# # print(d)
# d.update_realstate(2022, "June", 2000)
# d.update_salary(2022, "May", 20000)
# print(d)
# d.cash_flow()
# d.add_investment(2022, "October", 3000, 2000, 5000, 0)
# d.add_expenses(2022, "October", 3000, 2000, 1100, 500, 4000)
# print(d)
# d.net_worth()
# Maverick = Budget(2021)
# print(Maverick)
# Maverick.add_income(2021, "January", 2000, 300)
# Maverick.add_income(2021, "February", 2000, 300)
# Maverick.add_income(2021, "March", 2000, 300)
# Maverick.add_income(2021, "April", 2000, 300)
# Maverick.add_income(2021, "May", 2000, 300)
# Maverick.add_income(2021, "June", 2000, 300)
# Maverick.add_income(2021, "July", 2000, 300)
# Maverick.add_income(2021, "August", 2000, 300)
# Maverick.add_income(2021, "September", 2000, 300)
# Maverick.add_income(2021, "October", 2000, 300)
# Maverick.add_income(2021, "November", 2000, 300)
# Maverick.add_income(2021, "December", 2000, 300)
# Maverick.update_salary(2021, "November", 2300)
# Maverick.update_salary(2021, "December", 2300)
# Maverick.update_bonus(2021, "April", 400)
# Maverick.update_bonus(2021, "September", 400)
# Maverick.add_expenses(2021, "January", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "February", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "March", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "April", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "May", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "June", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "July", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "August", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "September", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "October", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "November", 400, 400, 100, 100, 300)
# Maverick.add_expenses(2021, "December", 400, 400, 100, 100, 300)
# Maverick.update_rent(2021, "November", 450)
# Maverick.update_rent(2021, "December", 450)
# Maverick.update_grocery(2021, "January", 600)
# Maverick.update_auto_expenses(2021, "February", 140)
# Maverick.update_medical_expenses(2021, "March", 120)
# Maverick.update_other_expenses(2021, "May", 500)
# Maverick.add_investment(2021, "January", 50, 10, 10, 20)
# Maverick.add_investment(2021, "February", 50, 10, 10, 20)
# Maverick.add_investment(2021, "March", 50, 10, 10, 20)
# Maverick.add_investment(2021, "April", 50, 10, 10, 20)
# Maverick.add_investment(2021, "May", 50, 10, 10, 20)
# Maverick.add_investment(2021, "June", 50, 10, 10, 20)
# Maverick.add_investment(2021, "July", 50, 10, 10, 20)
# Maverick.add_investment(2021, "August", 50, 10, 10, 20)
# Maverick.add_investment(2021, "September", 50, 10, 10, 20)
# Maverick.add_investment(2021, "October", 50, 10, 10, 20)
# Maverick.add_investment(2021, "November", 50, 10, 10, 20)
# Maverick.add_investment(2021, "December", 50, 10, 10, 20)
# Maverick.update_stocks(2021, "February", 100)
# Maverick.update_crypto(2021, "March", 100)
# Maverick.update_retirement_fund(2021, "April", 0)
# Maverick.update_realstate(2021, "December", 5)
# print(Maverick)
# Maverick.cash_flow()
# Maverick.net_worth()
```
